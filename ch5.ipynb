```python
def cross_entropy_error(y, t):
    return -np.sum(t * np.log(y + 1e7))
```

```python
def softmax(x):
    return np.exp(x - np.max(x)) / np.sum(x - np.max(x))
```

```python
def numerical_gradient(f, x):
    grad = np.zeros_like(x)
    
    for idx in range(x.size):
        tmp_val = x[idx]
        
        x[idx] = tmp_val + 1e-4
        fxh1 = f(x)
        
        x[idx] = tmp_val - 1e-4
        fxh2 = f(x)
        
        grad[idx] = (fxh1 - fxh2) / 2*(1e-4)
        
        x[idx]  = tmp_val
        
        return grad
```

```python
class MulLayer:
    def __init__(self):
        self.x = None
        self.y = None
        
    def forward(self, x, y):
        self.x = x
        self.y = y
        out = x * y
        return out
    
    def backward(self, dout):
        dx = dout * self.y
        dy = dout * self.x
        
        return dx, dy
    

```
